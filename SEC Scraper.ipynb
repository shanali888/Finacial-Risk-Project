# SEC Scraper for SEC Filings

```python
# load libraries
from selenium import webdriver
import pandas as pd
import numpy as np
```

```python
# load clean data, raw text file
sec = pd.read_csv('data-text-links.csv', low_memory=False)
sec = sec.drop(columns=['cik','fdate','wrdsfname','secpdate'])
```

### SEC Filings Scrapper

```python
# define data structure
sec_small = pd.DataFrame(sec.fname, columns=['fname'])
sec_small.loc[:,'idx'] = sec.index
sec_small.loc[:,'form'] = sec.form
sec_small = sec_small[sec_small.form != '8-K']
sec_small = sec_small.dropna().reset_index(drop=True)[:]
files, c, n = pd.DataFrame(), 1, sec_small.shape[0]
```

```python
# open driver & loop through filings
#files = pd.read_csv('sec_files.csv', low_memory=False)
driver = webdriver.Chrome('chromedriver_win32/chromedriver')
sec_small_temp = sec_small[(c-1):]
for idx, url in zip(sec_small_temp.idx, sec_small_temp.fname):
    # open page, extract, and save text  
    print(idx, url, c,'/',n)
    driver.get('https://www.sec.gov/Archives/' + url)
    try: 
        text = driver.find_element_by_tag_name('body').text
        files[idx] = [text]
    except: print('Missing File', idx)
    
    # save file dataframe periodically
    c += 1
    if (c%20000) == 0: break
    
# close page & process files output
driver.close()
```

```python
files.to_csv('sec_files.csv', index=False)
```
